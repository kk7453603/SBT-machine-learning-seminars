```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
plt.style.use('seaborn-colorblind')
%matplotlib inline
```

## Чтение данных

```python
use_cols = [
    'Pclass', 'Sex', 'Age', 'Fare', 'SibSp',
    'Survived'
]

data = pd.read_csv('./data/titanic.csv', usecols=use_cols)
```

```python
data.head(3)
```

## Типы колонок

```python
data.dtypes
```

## Статистики данных

```python
data.describe(include='all')
```

## Barplot

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Pclass',y='Survived',data=data)
```

## Countplot

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Pclass',data=data)
```

## Boxplot


```python
plt.figure(figsize=(15,10))
sns.boxplot(x='Pclass',y='Fare',data=data)
```

## Distplot (распредление непрерывной переменной)

```python
plt.figure(figsize=(15,10))
sns.distplot(a=data['Fare'],kde=False, bins=25)
```

## Scatter plot

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=data["Fare"],y=data["Pclass"],data=data)
```

## Матрица корреляции


```python
corrmat = data.corr()
fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(corrmat,cmap="YlGnBu",linewidths=.5,annot=True)
```

## Heatmap

```python
flights = sns.load_dataset("flights")
print(flights.head(5))
flights = flights.pivot("month", "year", "passengers")

fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(flights,cmap="YlGnBu",linewidths=.5,annot=True,fmt='d')
```
